```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# Load the model you just trained
model = YOLO('runs/detect/train5/weights/best.pt')
```

```python
# Path to your test image
img_path = '/home/wiebe/Desktop/Robotic/AE4317/testing_nn/data/pre-procest/316093264.jpg'

# Run inference


from PIL import Image

# Run inference
results = model(img_path, conf=0.25, imgsz=256) # Use 256 to avoid the stride warning

for r in results:
    # Get the image with bounding boxes drawn as a BGR numpy array
    im_bgr = r.plot() 
    
    # Convert BGR to RGB
    im_rgb = cv2.cvtColor(im_bgr, cv2.COLOR_BGR2RGB)
    
    # Use PIL to show the image directly
    display(Image.fromarray(im_rgb))
```
